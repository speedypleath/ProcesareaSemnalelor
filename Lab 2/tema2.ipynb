## 1. Frecventele emise de un contrabas se incadreaza intre 40Hz si 200Hz. Care este frecventa minima cu care trebuie esantionat semnalul trece-banda provenit din inregistrarea instrumentului, astfel incat semnalul discretizat sa contina toate componentele de frecventa pe care instrumentul le poate produce?

### $ f_{c} = \frac{200+40}{2} = 120Hz $

### $ B = \frac{200-40}{2} = 80Hz $

### $ f_{s} = f_{c} - B / 2 = 120 - 80 / 2 = 80Hz $

## 2. Calculati frecventa optima de esantionare sub-Nyquist pentru un semnal de banda B = 10Hz centrat in jurul lui $f_{c}$ = 90Hz pentru urmatoarele 3 valori posibile ale numarului de replici: a) m = 1, b) m = 2, c) m = 4

### Conditii:
### $\frac{2f_{c}-B}{m} \geq f_{s}  \geq \frac{2f_{c}-B}{m+1} <=> \frac{170}{m} \geq f_{s} \geq \frac{190}{m+1}$ 
### &emsp; &emsp; &emsp; &nbsp; &nbsp;  $f_{s}  \geq 2B$ &nbsp; &nbsp; &nbsp; $ <=>  f_{s} \geq 20$

|     $m$     | $(2f_{c}-B)/m$ | $(2f_{c}+B)/(m+1)$ |   $Optimum$   |
|    :----:   |     :----:     |       :----:       |    :----:     |
| 1           | 170 Hz         | 95 Hz              | 95 Hz         |
| 2           | 85 Hz          | 63.(3) Hz          | 63.(3) Hz     |
| 4           | 42.5 Hz        | 38 Hz              | 42.5 Hz       |

pt $f_{s} = 95Hz$ 90-95 = -5, -5+10/2 = 0, centrat in 0


```python
import numpy as np
import matplotlib.pyplot as plt

def x(t):
    return np.cos(20 * np.pi * t) # amplitudinea 1, frecventa 10, faza 0

atime = np.linspace(0, 1, 1000) # frecventa de esantionare 1000Hz intr-un orizont de timp de o secunda
plt.plot(atime, x(atime))
```

```python
X = np.fft.fft(x(atime))

xtime = np.linspace(0, 500, 500) # grila de frecvente [0, 500]
plt.plot(xtime, np.real(X[:len(X)//2]))
```
